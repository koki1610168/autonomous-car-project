# Importing necessary libraries

```python
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
import matplotlib.image as mpimg
```

```python
#Roading
img = mpimg.imread("test_images/straight_lines1.jpg")
#Converting the image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.title("Gray Scale")
plt.axis("off")
plt.imshow(gray, cmap="gray")
plt.show()
```

```python
print(img.shape)
```

# Try the canny edge detection

```python
canny = cv2.Canny(gray, 100, 200)
plt.title("Canny Edge Detection")
plt.axis("off")
plt.imshow(canny, cmap="gray")
plt.show()
```

```python
rec = cv2.rectangle(img, (200, 300), (1100, 650), (0, 255, 0), 5)
plt.imshow(rec)
plt.show()
```

# Use HSV to isolate white or yellow lines

```python
rec = cv2.rectangle(img, (200, 300), (1100, 650), (255, 255, 255), 2)
plt.imshow(rec)
plt.show()
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)


plt.title("HSV Isolation")

plt.imshow(img_hsv, cmap="gray")
plt.show()
```
